```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax1.set_ylabel('Z')
ax1.set_xlabel('X')

ax2.plot(x,y)
ax2.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_xlim(20,22)
ax2.set_xlim(30,50)
```

```python
plt.subplots(nrows=1,ncols=2)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(14,2))
ax[0].plot(x,y,'b--',linewidth=3)
ax[1].plot(x,z,'r',linewidth=3)
```
